## AGM - Affiliation Graph Model

### AGM is a model used to generate a graph/network
#### 1. V = set of nodes
#### 2. C = set of communities
#### 3. M = relationship->tell the nodes in communities
#### 4. wrong! --> Pc = probability of generating a connection in a community;  Modified: Pc = is a generation prob for a community
#### 5. for pair(u,v), P(u,v) is the prob to generate a connection by Pc

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
# community class definition
class community:
    def __init__(self, member, prob):
        self.member = member
        self.prob = prob
```

```python
from itertools import combinations
from random import random
```

```python
def AGM(nodes, communities):
    # create an empty graph
    G = nx.Graph()
    # add nodes to this graph
    G.add_nodes_from(nodes)
    
    # generate links within comminities
    for c in communities:
        for nodePairs in combinations(c.member, 2):
            if random() <= c.prob:
                G.add_edge(nodePairs[0], nodePairs[1])
    return G
```

```python
# example graph with two communities (make sure two communities have
# overlapping nodes)
nodes = range(12)
communities = [community(range(5), 0.7), community(range(4,12), 0.8)]
G = AGM(nodes, communities)
nx.draw(G)
```

```python
# example graph with two communities
nodes = range(12)
communities = [community([0,1,2,3,4,5,6,7], 0.8), community([7,8,9,10,11], 0.8)]
G = AGM(nodes, communities)
nx.draw(G)
```

```python
# example graph with three communities
nodes = range(24)
communities = [community(range(12), 0.8),
               community(range(6,17), 0.8),
               community(list(range(8,12))+list(range(17,24)), 0.8)]
G = AGM(nodes,communities)
nx.draw(G)
```
