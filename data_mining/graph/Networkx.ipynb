### Learn how to use Networkx

```python
import networkx as nx
```

#### Node

```python
# create a new empty graph
# 根据图的定义，一个图包含一个节点集合和一个边集。在NetworkX中，节点可以是任何哈希对象
G = nx.Graph()
```

```python
# add a node from a node
G.add_node(1)

# add nodes from a node list
G.add_nodes_from([2,3,4,5])
```

```python
nx.draw(G)
```

```python
# add nodes from any iterable containers[nbunch - node bunch] of nodes (list, set, graph, file ...)
H = nx.path_graph(10)
G.add_nodes_from(H)
```

```python
# H is a graph
H
```

```python
nx.draw(H)
```

```python
nx.draw(G)
```

```python
# Here, use graph--H as the node of graph--G
G = nx.Graph()
G.add_node(H)
```

```python
# Now, the node represents the graph H
nx.draw(G)
```

#### Edge

```python
# initialize a new graph
G_2 = nx.Graph()
```

```python
G_2.add_edge(1,2)
# or
e = (1,2)
G_2.add_edge(*e)
```

```python
nx.draw(G_2)
```

```python
G_2.add_edges_from([(1,2), (1,3)])
```

```python
nx.draw(G_2)
```

```python
# Add edges from ebunch(An ebunch-edge bunch is any iterable container of edge-tuples)
## An edge-tuple can be a 2-tuple of nodes or a 3-tuple with 2 nodes followed by an edge attribute dictionary, 
## e.g., (2, 3, {'weight': 3.1415})

# add edges from H
G_2.clear()
G_2.add_edges_from(H.edges)
```

```python
nx.draw(G_2)
```

#### 快速遍历每一条边，使用邻接迭代器实现

```python
FG = nx.Graph()
FG.add_weighted_edges_from([(1,2,0.125), (1,3,0.75), (2,4,1.2),(3,4,0.375)])
```

```python
nx.draw(FG)
```

```python
for n, nbrs in FG.adjacency():
    for nbr, eattr in nbrs.items():
        data = eattr['weight']
        if data < 0.5:
            print(n, nbr, data)
```
